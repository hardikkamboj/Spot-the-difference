```python
import cv2
import numpy as np 

import matplotlib.pyplot as plt 
```

```python
im1 = cv2.imread('diff1.png')         
im2 = cv2.imread('diff2.png') 
```

```python
gray1 = cv2.cvtColor(im1, cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(im2, cv2.COLOR_BGR2GRAY)
```

```python
orb = cv2.ORB_create(50)  
```

```python
kp1, des1 = orb.detectAndCompute(gray1, None)  #kp1 --> list of keypoints
kp2, des2 = orb.detectAndCompute(gray2, None)
```

```python
result1 = cv2.drawKeypoints(im1, kp1, None)
result2 = cv2.drawKeypoints(im2, kp2, None)
```

```python
plt.figure(figsize = (24,8))
plt.subplot(121)
plt.imshow(result1[:,:,::-1])
plt.axis('off')

plt.subplot(122)
plt.imshow(result2[:,:,::-1])
plt.axis('off')
```

```python
matcher = cv2.DescriptorMatcher_create(cv2.DESCRIPTOR_MATCHER_BRUTEFORCE_HAMMING)

# Match descriptors.
matches = matcher.match(des1, des2, None) 
```

```python
matches = sorted(matches, key = lambda x:x.distance)
```

```python
result3 = cv2.drawMatches(im1,kp1, im2, kp2, matches[:10], None)
```

```python
plt.figure(figsize = (10,8))
plt.imshow(result3[:,:,::-1])
plt.axis('off')
```

```python
points1 = np.zeros((len(matches), 2), dtype=np.float32)  #Prints empty array of size equal to (matches, 2)
points2 = np.zeros((len(matches), 2), dtype=np.float32)
```

```python
for i, match in enumerate(matches):
   points1[i, :] = kp1[match.queryIdx].pt    #gives index of the descriptor in the list of query descriptors
   points2[i, :] = kp2[match.trainIdx].pt    #gives index of the descriptor in the list of train descriptors
```

```python
h, mask = cv2.findHomography(points1, points2, cv2.RANSAC)
 
  # Use homography
height, width, channels = im2.shape
im1Reg = cv2.warpPerspective(im1, h, (width, height))  #Applies a perspective transformation to an image.
```

```python
plt.figure(figsize = (10,8))
plt.imshow(im1Reg[:,:,::-1])
plt.axis('off')
```

```python
plt.figure(figsize = (36,8))
plt.subplot(131)
plt.imshow(im1[:,:,::-1])
plt.axis('off')

plt.subplot(132)
plt.imshow(im2[:,:,::-1])
plt.axis('off')

plt.subplot(133)
plt.imshow(im1Reg[:,:,::-1])
plt.axis('off')
```

```python
sub1 = cv2.subtract(im1, im2)
sub2 = cv2.subtract(im1Reg, im2)
```
